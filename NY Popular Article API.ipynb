```python
import json
import requests
import pandas as dataframe
import pandas as pd
```

```python
url = 'http://api.nytimes.com/svc/mostpopular/v2/mostviewed/business/7.json?api-key=d50955abe9630b297dd92ebab442b122%3A8%3A73272974'
```

```python
r = requests.get(url)
```

```python
print (r.text)
```

```python
nyarticles = json.loads(r.text)
```

```python
print (nyarticles["results"])
```

```python
urls=[]
columns=[]
bylines=[]
titles=[]
abstracts=[]

for item in nyarticles['results']:
    print (item["url"])
    urls.append(item["url"])
    print (item["column"])
    columns.append(item["column"])
    print (item["byline"])
    bylines.append(item["byline"])
    print (item["title"])
    titles.append(item["title"])
    print (item["abstract"])
    abstracts.append(item["abstract"])
```

```python
columns = {'URL': urls, 'Column': columns, 'By': bylines, 'Title': titles, "Abstract": abstracts}
df = pd.DataFrame(columns)
df
```
